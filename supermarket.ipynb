# Q11-Which product line helps you generate the most income?

```python
#making barplot
pro=df[['Product line','gross income']].groupby(['Product line'],as_index=False).sum().sort_values(by='gross income',ascending=False)
plt.figure(figsize=(20,8))
sns.barplot(x='Product line',y='gross income',data=pro)
plt.title('Gross income vs product line')
```

A11-Food and beverages generated the most income

# Q13-Which day of the week has maximum sales?

```python
#First we convert date from object to datetimeobject
df['Date']=pd.to_datetime(df['Date'])
```

```python
#return daynames from the dates
df['weekday']=df['Date'].dt.day_name()
```

```python
#set index to date
df.set_index('Date', inplace=True)
```

```python
df.head()
```

```python
#making countplot for weekday
plt.figure(figsize=(8,6))
plt.title('Daily Sales by Day of the Week')
sns.countplot(x=df['weekday'])
```

A13-Saturday has maximum sales.Also, Monday has least sales, so they can put offers on Monday to increase sales

# Q14-Which hour of the day is busiest?

```python
#Converting time object to datetime object
df['Time']=pd.to_datetime(df['Time'])
df['Hour']=(df['Time']).dt.hour
df['Hour'].unique()
```

```python
#making lineplot
sns.lineplot(x='Hour',y='Quantity',data=df).set_title('Product sales per hour')
```
